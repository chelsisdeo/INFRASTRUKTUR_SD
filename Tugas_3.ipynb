### Diberikan data test untuk kasus unguided praktikum minggu ini, lakukan identifikasi pada data tersebut apakah terdapat perbedaan karakteristik data train pada praktikum sebelumnya.

```python
import pandas as pd

# Baca file CSV train dan test
train_data = pd.read_csv('train_features (1).csv')
test_data = pd.read_csv('test_features.csv')

# Tampilkan beberapa baris pertama dari kedua dataset
print("Data Train:")
print(train_data.head())
print("\nData Test:")
print(test_data.head())

# Identifikasi kolom yang ada di kedua dataset
print("\nKolom di Train:", train_data.columns)
print("\nKolom di Test:", test_data.columns)

# Informasi dasar tentang kedua dataset (missing values, tipe data, dll.)
print("\nInformasi Data Train:")
print(train_data.info())

print("\nInformasi Data Test:")
print(test_data.info())

# Periksa perbedaan statistik deskriptif
print("\nStatistik deskriptif Data Train:")
print(train_data.describe())

print("\nStatistik deskriptif Data Test:")
print(test_data.describe())

# Cek jumlah missing value di kedua dataset
print("\nMissing value di Data Train:")
print(train_data.isnull().sum())

print("\nMissing value di Data Test:")
print(test_data.isnull().sum())

# Periksa apakah ada perbedaan antara jumlah kolom atau nama kolom
if set(train_data.columns) == set(test_data.columns):
    print("\nKolom di kedua dataset sama.")
else:
    print("\nTerdapat perbedaan kolom antara train dan test.")

# Membandingkan nilai rata-rata dari setiap kolom (hanya untuk kolom numerik)
mean_diff = train_data.mean() - test_data.mean()
print("\nPerbedaan nilai rata-rata antara Data Train dan Test:")
print(mean_diff)
```

```python
import pandas as pd

# Baca file CSV train dan test
train_data = pd.read_csv('train_features (1).csv')
test_data = pd.read_csv('test_features.csv')

# Tampilkan beberapa baris pertama dari kedua dataset
print("Data Train:")
print(train_data.head())
print("\nData Test:")
print(test_data.head())

# Identifikasi kolom yang ada di kedua dataset
print("\nKolom di Train:", train_data.columns)
print("\nKolom di Test:", test_data.columns)

# Informasi dasar tentang kedua dataset (missing values, tipe data, dll.)
print("\nInformasi Data Train:")
print(train_data.info())

print("\nInformasi Data Test:")
print(test_data.info())

# Periksa perbedaan statistik deskriptif
print("\nStatistik deskriptif Data Train:")
print(train_data.describe())

print("\nStatistik deskriptif Data Test:")
print(test_data.describe())

# untuk mengecek jumlah missing value di kedua dataset
print("\nMissing value di Data Train:")
print(train_data.isnull().sum())

print("\nMissing value di Data Test:")
print(test_data.isnull().sum())

# memeriksa apakah ada perbedaan antara jumlah kolom atau nama kolom
if set(train_data.columns) == set(test_data.columns):
    print("\nKolom di kedua dataset sama.")
else:
    print("\nTerdapat perbedaan kolom antara train dan test.")

# Membandingkan nilai rata-rata dari setiap kolom (hanya untuk kolom numerik)

numeric_cols = train_data.select_dtypes(include=['number']).columns.intersection(test_data.select_dtypes(include=['number']).columns)
mean_diff = train_data[numeric_cols].mean() - test_data[numeric_cols].mean()
# --- Akhir perubahan ---

print("\nPerbedaan nilai rata-rata antara Data Train dan Test:")
print(mean_diff)
```
